```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10,0.5)
y = x ** 2
```

# Scatter Plot

```python
plt.scatter(x,y)
plt.scatter(x,x**3)
plt.show()
```

# Line Graph

```python
plt.plot(x,y,'r--',linewidth = 1.5,label="fun1")
plt.plot(x,x**3,'b-',label="fun2")
plt.axis([0,20,0,1000])
plt.xlabel("Numbers")
plt.ylabel("Square")
plt.grid()
plt.legend()
plt.title("Test")
plt.text(8,400,"Hello",fontsize=12)
plt.show()
```

# Bubble Chart

```python
# Special case of scaller plot where the input argument s "size" is a variable array
# Used when we wish to show multiple dependent variables on the same graph
# adding an extra dim without going 3d
```

```python
x = np.arange(0,10)
y = np.arange(20,30)
z = np.arange(50,150,10)
```

```python
plt.scatter(x,y,s=z)
plt.show()
```

```python
# Changing the marker
plt.scatter(x,y,s=z,marker="^")
plt.show()
```

```python
# Adding variable colors
"c parameter could be an array of the same size"
color = np.arange(1,11)
plt.scatter(x,y,s=z,marker="^",c = color)
plt.show()
```

```python
# Adding text at each location of the graph
color = np.arange(1,11)
plt.scatter(x,y,s=z,marker="^",c = color)

for i in range(len(x)):
  plt.text(x[i]+0.1,y[i]+.01,z[i])
plt.show()
```

# Pie Chart

```python
sizes = [3,4,6,2]
colors = ['blue','purple','red','pink']
labels = ["A","B","C","D"]
# To show  percentages use autopct, %% for adding percentage to the end
plt.pie(sizes,colors = colors,labels = labels,autopct = "%.2f%%")
plt.axis("equal")  # to make it round
plt.show()
```

```python
# using explode to make a slice of pie

sizes = [3,4,6,2]
colors = ['blue','purple','red','pink']
labels = ["A","B","C","D"]
explode = [0.1,0,0,0]
# To show  percentages use autopct, %% for adding percentage to the end
plt.pie(sizes,colors = colors,labels = labels,autopct = "%.2f%%",explode= explode)
plt.axis("equal")  # to make it round
plt.show()
```

# Histrogram

```python
# Use to check the frequency of data
# each tower is called bin
a = [1,2,3,4,1,2,3,6,5,1,2,3,5,2,3,6,5,23,5,2,1,5,4,5]
# bins could be array also
plt.hist(a,bins = 100,edgecolor ="black")
plt.xticks(np.arange(1,30))
plt.show()
```

# Bar graphs

```python
# Generally when we have low sample points then it provides a good representation
x = np.arange(0,10)
y = np.arange(20,30)
z = np.arange(50,150,10)
```

```python
plt.bar(y,z,width = 0.3)
plt.show()
```

```python
# x have some strings
x = ["2011-0", "2011-1", "2011-2", "2011-3", "2011-4", "2011-5", "2011-6", "2011-7", "2011-8", "2011-9"]
plt.bar(x,z,width = 0.3)
plt.show()
```

```python
# use xticks -> rotation
x = ["2011-0", "2011-1", "2011-2", "2011-3", "2011-4", "2011-5", "2011-6", "2011-7", "2011-8", "2011-9"]
plt.bar(x,z,width = 0.3)
plt.xticks(rotation = 40)
plt.show()
```
